```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to load
filepath_city = "C:/Users/Deepti/DataAnalytics/Week 5 - Matplotlib/Homework/Instructions/Pyber/data/city_data.csv"
filepath_ride = "C:/Users/Deepti/DataAnalytics/Week 5 - Matplotlib/Homework/Instructions/Pyber/data/ride_data.csv"

newfile_city = pd.read_csv(filepath_city)
newfile_ride = pd.read_csv(filepath_ride)

```

```python
#Merged the data into a single dataset
mergedfile = pd.merge(newfile_ride, newfile_city, on = "city", how = "left")
mergedfile
```

```python
# Build a [Bubble Plot] that showcases the relationship between four key variables:
# * Average Fare ($) Per City
# * Total Number of Rides Per City
# * Total Number of Drivers Per City
# * City Type (Urban, Suburban, Rural)

# Step 1: #Getting Data from Urban Areas
urban_pop = mergedfile.loc[mergedfile["type"]=="Urban"]
#print(urban_pop)
avg_fare_urban =urban_pop.groupby(["city"])[["fare"]].mean()
tot_ride_urban = urban_pop.groupby(["city"])[["ride_id"]].count()
tot_driver_urban = urban_pop.groupby(["city"])[["driver_count"]].mean()
#print(tot_driver_urban)

# Step 2 :Getting Data from Suburban Areas
suburban_pop = mergedfile.loc[mergedfile["type"]=="Suburban"]
avg_fare_suburban =suburban_pop.groupby(["city"])[["fare"]].mean()
tot_ride_suburban = suburban_pop.groupby(["city"])[["ride_id"]].count()
tot_driver_suburban = suburban_pop.groupby(["city"])[["driver_count"]].mean()
#print(suburban_pop)

# Step 3 : Getting Data from Rural Areas
rural_pop = mergedfile.loc[mergedfile["type"]=="Rural"]
avg_fare_rural =rural_pop.groupby(["city"])[["fare"]].mean()
tot_ride_rural = rural_pop.groupby(["city"])[["ride_id"]].count()
tot_driver_rural = rural_pop.groupby(["city"])[["driver_count"]].mean()
#print(rural_pop)

# Step 4: Plotting in Scatter...
plt.scatter(tot_ride_urban, avg_fare_urban, s=tot_driver_urban*10, marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.5, label="Urban")
plt.scatter(tot_ride_suburban, avg_fare_suburban, s=tot_driver_suburban*10, marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.5, label="Suburban")
plt.scatter(tot_ride_rural, avg_fare_rural, s=tot_driver_rural*10, marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.5, label="Rural")

plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare($)")
plt.legend({"Urbal","Suburban","Rural"})
plt.text(8,2,"Note: Circle Size correlates with driver count per city.", fontsize=12,verticalalignment='top', horizontalalignment='left')
plt.show()
```

```python
# 1. Pie Chart: % of Total Fares by City Type

# Step 1 :Calculating required values
total_fare_urban= urban_pop["fare"].sum()
#total_fare_urban
total_fare_suburban= suburban_pop["fare"].sum()
#total_fare_suburban
total_fare_rural= rural_pop["fare"].sum()
#total_fare_rural
overall_fare =total_fare_urban + total_fare_suburban + total_fare_rural
#overall_fare
per_urban =total_fare_urban /overall_fare * 100
#per_urban
per_suburban =total_fare_suburban /overall_fare * 100
#per_suburban
per_rural =total_fare_rural /overall_fare * 100
#per_rural

# Step 2 : Making a pie chart.
x = [per_urban, per_suburban, per_rural]
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "Gold"]
explode = [0.1,0,0]
plt.pie(x, explode=explode, labels=labels,colors=colors,autopct="%1.01f%%",shadow=True,startangle=140)
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
# 2. Pie Chart: % of Total Rides by City Type

# Step 1 :Calculating required values
urban_ride = urban_pop["ride_id"].count()
#urban_ride
suburban_ride = suburban_pop["ride_id"].count()
#suburban_ride
rural_ride = rural_pop["ride_id"].count()
#rural_ride
total_ride = urban_ride +rural_ride+ suburban_ride
#total_ride
per_urban_ride = urban_ride /total_ride * 100
#per_urban_ride
per_suburban_ride = suburban_ride /total_ride * 100
#per_suburban_ride
per_rural_ride = rural_ride /total_ride * 100
#per_rural_ride


# Step 2 : Making a pie chart.
y = [per_urban_ride, per_suburban_ride, per_rural_ride]
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "Gold"]
explode = [0.1,0,0]
plt.pie(y, explode=explode, labels=labels,colors=colors,autopct="%1.01f%%",shadow=True,startangle=140)
plt.title("% of Total Rides by City Type")
plt.show()

```

```python
# 3. Pie Chart: % of Total Drivers by City Type

# Step 1 :Calculating required values
total_cal=newfile_city.groupby(["type"])[["driver_count"]].sum()
total_cal


#Step 2 : Making a pie chart.
labels = ["Rural", "Suburban", "Urban"]
colors = ["Gold", "lightskyblue","lightcoral"]
explode = [0.1,0,0]
plt.pie(total_cal, explode=explode, labels=labels,colors=colors,autopct="%1.01f%%",shadow=True,startangle=140)
plt.title("% of Total Drivers by City Type")
plt.show()
```
